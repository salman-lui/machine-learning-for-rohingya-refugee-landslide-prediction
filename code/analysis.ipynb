```python
import pandas as pd
import os
```

```python
train_test_data = pd.read_csv(os.path.join('..','data','Landslide_Data.csv'))
point_data = pd.read_csv(os.path.join('..','data','Point_data.csv'))
```

```python
for col in ['Geology', 'LULC_2020',  'SoilTexture', 'SoilType']:
    train_test_data[col] = train_test_data[col].astype('category')
    point_data[col] = point_data[col].astype('category')
   
       
```

```python
print('In train test data:\n')
for col in ['Geology', 'LULC_2020',  'SoilTexture', 'SoilType']:
    print('Unique categories for '+col)
    print(list(train_test_data[col].unique()))
    print('\n')
    
```

```python
print('In point data:\n')
for col in ['Geology', 'LULC_2020',  'SoilTexture', 'SoilType']:
    print('Unique categories for '+col)
    print(list(point_data[col].unique()))
    print('\n')
    
```

```python
print('Missing categories (which are presesnt in point data but not in train test data):\n')
for col in ['Geology', 'LULC_2020',  'SoilTexture', 'SoilType']:
    print('Missing categories for '+col)
    ttd=list(train_test_data[col].unique())
    pd=list(point_data[col].unique())
    m = list(set(pd) - set(ttd))
    print(m)
    print('\n')
    
```
